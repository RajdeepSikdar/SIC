```python
def fact(n):
    if(n==1 or n==0):
       return 1
    return fact(n-1)*(n)
print(fact(5))
```

```python
def fibonaci(n):
    if(n==1 or n==0):
      return 1
    return fibonaci(n-1)+fibonaci(n-2)
print(fibonaci(9))
```

```python
#create a random matrix of 14 * 14 having random values between 0 and 255 and show image view of the matrix and shorten the immage by max pooling
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import block_reduce
matrix = np.random.randint(0, 256, (14,14), dtype=np.uint8)
plt.figure(figsize=(6, 3))
plt.subplot(1, 2, 1)
plt.imshow(matrix, cmap='gray', vmin=0, vmax=255)
plt.title('Original Image')
plt.axis('off')
pooled = block_reduce(matrix, block_size=(2, 2), func=np.max)
plt.subplot(1, 2, 2)
plt.imshow(pooled, cmap='gray', vmin=0, vmax=255)
plt.title('Max Pooled Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import cv2
import os
import sys
video_path = r'falling-ball-led-light-ball-satisfying-music-shorts-viral-trending-youtubeshorts-funny-ytshorts.savetube.me.mp4'
if not os.path.isfile(video_path):
    print(f"Error: The file '{video_path}' does not exist.")
    sys.exit()
cap = cv2.VideoCapture(video_path)
frames = []
while True:
    ret, frame = cap.read()
    if not ret:
        break
    frames.append(frame)
cap.release()
if not frames:
    print("Error: No frames were captured from the video.")
    sys.exit()
frames.reverse()
height, width, layers = frames[0].shape
fps = 30  
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
output_path = 'reversed_videos.mp4'
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))
for frame in frames:
    out.write(frame)
out.release()
print(f"Reversed video saved to '{output_path}'")
```
